```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
data = pd.read_csv("/content/drive/MyDrive/email_spam.csv", encoding= 'latin-1')
data.head()
```

```python
data.tail()
```

```python
# Selecting only the 'type' and 'text' columns
spam_data = data[['type', 'text']]

# Display the first few rows of the new dataset
spam_data.head()
```

```python
#  Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(spam_data['text'], spam_data['type'], test_size=0.2, random_state=42)

```

```python
# Display the shapes of the training and test sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)

```

```python
# Train the Naive Bayes Model
# Use CountVectorizer to convert text data into numerical features
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Train the Naive Bayes model
spam_classifier = MultinomialNB()
spam_classifier.fit(X_train_vectorized, y_train)


```

```python
#  Evaluate the Model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Transform the test set using the same vectorizer
X_test_vectorized = vectorizer.transform(X_test)

# Make predictions on the test set
predictions = spam_classifier.predict(X_test_vectorized)

# Evaluate the model using metrics
accuracy = accuracy_score(y_test, predictions)
classification_rep = classification_report(y_test, predictions)
confusion_mat = confusion_matrix(y_test, predictions)

# Display the evaluation metrics
print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_rep)
print("\nConfusion Matrix:\n", confusion_mat)
```
